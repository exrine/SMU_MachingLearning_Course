## 2. Use loop to obtain mutiple means and variances

```python
n = np.arange(100, 100001, 100)
MEAN = np.empty_like(n,dtype=float)   # be very careful about "dtype", what if we omit it?
VAR = np.empty_like(n,dtype=float)
for i, j in enumerate(n):
    x = np.random.normal(size=j)
    MEAN[i] = np.mean(x)
    VAR[i] = np.var(x)
```

## 3. Plot figures

```python
# plot in the same figure
import matplotlib.pyplot as plt
plt.plot(n,MEAN,'-',n,VAR,'--')
plt.xlabel('Sample Size')
plt.ylabel('Value')
plt.title('Plot Mean and Variance in the Same Figure')
plt.xlim(n[0],n[-1])
plt.grid()
plt.legend(['Mean','Variance'],loc='best');   # what does "loc='best'" stand for?
```

```python
# plot in the different subfigures
plt.subplot(2,1,1)
plt.plot(n,MEAN,'b')
plt.xlabel('Sample Size')
plt.ylabel('Value')
plt.title('(a) Mean Plot')
plt.xlim(n[0],n[-1])
plt.ylim(-0.2,0.2)
plt.grid()

plt.subplot(2,1,2)
plt.plot(n,VAR,'r')
plt.xlabel('Sample Size')
plt.ylabel('Value')
plt.title('(b) Variance Plot')
plt.xlim(n[0],n[-1])
plt.ylim(0.9,1.1)
plt.grid()

plt.tight_layout()     # what will happen if we remove "tight_layout"?
```
